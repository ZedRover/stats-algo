1(a)
----
$H_0$ : 样本数据的均值与3.1差异为0

```python
x = stats.norm(3,1)
data = x.rvs(500)
tval = (3.1-np.mean(data))/stats.sem(data)
td = stats.t(len(data)-1)
p = 2*td.sf(tval)
print(p)
if p< 0.05:
    print("reject H0")
else:
    print("accept H0")
td = stats.t(4)
td.pdf(1-0.0125/2)
```

$\because$ 样本方差均值未知，已知样本均值和总体均值的标准化差异

$\therefore$ 构造t统计量

$\because$ 在此次检验中 p = 0.007 < 0.05

$\therefore$ 拒绝$H_0$ , 样本均值不为3.1

$H_0$ : 骰子是公平的


```python
n = 235
p = 1/6
bd = stats.binom(n,p)
pVal1 = bd.sf(50)
pVal2 = stats.binom_test(51,n,p)
if pVal1 >= 0.05:
    print("p=",round(pVal1,3),">=0.05"," 单边检验接受H0")
else:
    print("p=",round(pVal1,3),"<0.05"," 单边检验拒绝H0")

if pVal2>=0.05:
    print("p=",round(pVal2,3),">=0.5"," 双边检验接受H0")
else:
    print("p=",round(pVal2,3) ,"<0.05"," 双边检验拒绝H0")
```
