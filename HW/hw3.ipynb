```python
import pandas as pd
import numpy as np 
from scipy import stats
```

1(a)
----
$H_0$ : 样本数据的均值与3.1差异为0

```python
np.random.seed(1000)
x = stats.norm(3,1)
data = x.rvs(500)
zval = (3.1-np.mean(data))/(1/np.sqrt(500))
p = 2*stats.norm.sf(zval)
print(p)
if p< 0.025:
    print("reject H0")
else:
    print("accept H0")
```

$\because$ 样本数据较大，已知方差

$\therefore$ 构造z统计量

$\because$ 在此次检验中 p = 0.0045 < 0.05

$\therefore$ 拒绝 $H_0$ , 样本均值不为3.1

$H_0$ : 骰子是公平的


```python
n = 235
p = 1/6
bd = stats.binom(n,p)
pVal1 = bd.sf(50)
pVal2 = stats.binom_test(51,n,p)
if pVal1 >= 0.05:
    print("p=",round(pVal1,3),">=0.05"," 单边检验接受H0")
else:
    print("p=",round(pVal1,3),"<0.05"," 单边检验拒绝H0")

if pVal2>=0.05:
    print("p=",round(pVal2,3),">=0.05"," 双边检验接受H0")
else:
    print("p=",round(pVal2,3) ,"<0.05"," 双边检验拒绝H0")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set("poster",'ticks','muted')
def showpic(axis,data):
    axis.hist(data ,bins = 50,alpha= 0.5)
    axis.set_xticks([1,0.5,1])
data = np.random.random(1000000)
fig,axs = plt.subplots(1, 3)
fig.set_size_inches(15,10)
showpic(axs[0],data)
showpic(axs[1],np.mean(data.reshape(1000000//2,2),axis =1))
showpic(axs[2],np.mean(data.reshape(1000000//10,10),axis=1))
plt.tight_layout()
plt.show()
   
```
